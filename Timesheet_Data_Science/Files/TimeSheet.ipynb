```python
# Univariate Analysis
# Define shades of blue colors
blue_colors = ['#1f77b4', '#3498db', '#6fa8dc', '#a6c7e2', '#c9e4f7']

# Calculate the number of projects in each project type
project_type_counts = df['Project_Type'].value_counts()

# Create a pie chart with shades of blue colors
plt.figure(figsize=(6, 6))
plt.pie(project_type_counts, labels=project_type_counts.index, autopct='%1.1f%%', startangle=140, colors=blue_colors)
plt.title('Distribution of Projects by Project Type')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
#Univariate Analysis
# Calculate the number of tasks
task_counts = df['Task'].value_counts()

# Sort the counts in descending order
task_counts = task_counts.sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(10, 6))  # Adjust the size as needed
task_counts.plot(kind='bar')
plt.title('Number of Tasks')
plt.xlabel('Task')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
#Bivariate Analysis
# Define shades of blue and light blue colors
blue_colors = ['#1f77b4', '#3498db', '#6fa8dc', '#a6c7e2', '#c9e4f7']
light_blue_colors = ['#add8e6', '#87ceeb', '#87cefa', '#00bfff', '#00ced1']

# Group by project type and calculate total hours for each type
total_hours_by_project_type = df.groupby('Project_Type')['Total_Hours'].sum()

# Create a bar chart with shades of blue and light blue colors
plt.figure(figsize=(10, 6))
total_hours_by_project_type.plot(kind='bar', color=blue_colors + light_blue_colors)
plt.title('Total Hours by Project Type')
plt.xlabel('Project Type')
plt.ylabel('Total Hours')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, classification_report

# Load your dataset
data = pd.read_csv('Dataset.csv')

# Define features and target variable
features = ['Project_Type','Project_Name','Mon_Hours', 'Tues_Hours', 'Wed_Hours', 'Thur_Hours', 'Fri_Hours', 'Sat_Hours', 'Sun_Hours']
target = 'Total_Hours'

# Split data into features and target variable
X = data[features]
y = data[target]

# Convert Total_hours to binary labels (<35 hours or >=35 hours)
y_binary = (y < 35).astype(int)

# Initialize and train the Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_classifier.fit(X, y_binary)

# Apply the model to predict which users will have total hours less than 35 in the future
predictions = gb_classifier.predict(X)

# Calculate and print the accuracy
accuracy = accuracy_score(y_binary, predictions)
print("Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(y_binary, predictions))

# Get the list of users predicted to have total hours less than 35
users_less_than_35_hours = data[predictions == 1]

# Remove duplicate entries
unique_users_less_than_35_hours = users_less_than_35_hours.drop_duplicates(subset=['User_Name'])

# Save the list of users to a CSV file
unique_users_less_than_35_hours.to_csv('User_List_GradientBoosting.csv', index=False)
```

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

# Load your dataset
data = pd.read_csv('Dataset.csv')

# Define features and target variable
features = ['Project_Type','Project_Name','Mon_Hours', 'Tues_Hours', 'Wed_Hours', 'Thur_Hours', 'Fri_Hours', 'Sat_Hours', 'Sun_Hours']
target = 'Total_Hours'

# Split data into features and target variable
X = data[features]
y = data[target]

# Convert Total_hours to binary labels (<35 hours or >=35 hours)
y_binary = (y < 35).astype(int)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

# Initialize and train the Support Vector Machine (SVM) Classifier
svm_classifier = SVC(kernel='linear', random_state=42)  # You can experiment with different kernels
svm_classifier.fit(X_train, y_train)

# Make predictions on the test set
predictions = svm_classifier.predict(X_test)

# Calculate and print the accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

# Print the classification report
print("\nClassification Report:")
print(classification_report(y_test, predictions))

# Apply the model to predict which users will have total hours less than 35 in the future
predictions = svm_classifier.predict(X)

# Get the list of users predicted to have total hours less than 35
users_less_than_35_hours = data[predictions == 1]

# Remove duplicate entries
unique_users_less_than_35_hours = users_less_than_35_hours.drop_duplicates(subset=['User_Name'])

# Save the list of users to a CSV file
unique_users_less_than_35_hours.to_csv('User_List_SVM.csv', index=False)
```

```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

# Load your dataset
data = pd.read_csv('Dataset.csv')

# Define features and target variable
features = ['Mon_Hours', 'Tues_Hours', 'Wed_Hours', 'Thur_Hours', 'Fri_Hours', 'Sat_Hours', 'Sun_Hours']
target = 'Total_Hours'

# Split data into features and target variable
X = data[features]
y = data[target]

# Convert Total_hours to binary labels (<35 hours or >=35 hours)
y_binary = (y < 35).astype(int)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

# Initialize and train the Multilayer Perceptron (MLP) Classifier
mlp_classifier = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)
mlp_classifier.fit(X_train, y_train)

# Make predictions on the test set
predictions = mlp_classifier.predict(X_test)

# Calculate and print the accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

# Print the classification report
print("\nClassification Report:")
print(classification_report(y_test, predictions))

# Apply the model to predict which users will have total hours less than 35 in the future
predictions = mlp_classifier.predict(X)

# Get the list of users predicted to have total hours less than 35
users_less_than_35_hours = data[predictions == 1]

# Remove duplicate entries
unique_users_less_than_35_hours = users_less_than_35_hours.drop_duplicates(subset=['User_Name'])

# Save the list of users to a CSV file
unique_users_less_than_35_hours.to_csv('User_List_MLP.csv',index=False)
```
